# Richiami sul Modello Lineare Classico

- $Y$ - v.quantitativa dipendente
- $X_1..X_k$ - k-v indipendenti/ esplicative quantitative


Obietivo: construire un modello che descriva la dipendenza da Y dalle k-var $X_j$

$$Y = f(X_1...X_k) + U$$

- $U$- var non osservabile - esprime componente residuale che lega Y a k- var $X_j$. SI ipotizza - LINEARE

Varie forme del modello:

1. A livello concettuale:
$$Y = \beta_0+ \beta_1X_1 + ...+ \beta_kX_k + U$$

- $\beta_j$ - parametri ignoti oggetto di stima.

2. A livello campionario, prima di effetturare le estrazioni: mediante campionamento casuale semplice-ccs

- $X_j$ - noti(pre-fissati) ,
- $Y$ - dipende dalle us che entranno nel campione
- possibili risultati sono descritti n (k+1)-uple

$$(Y, x_{11}, x_{1k})...(Y_n, x_{n1}..x_{nk})$$

- $Y_i$ - natura stocastica (aleatoria, casuale), derivano dalle estrazione
- $X_j$- sono deterministici

Modello specificato mediane n-equazioni , una per ciascuna unita, insieme ad una seria di ipotesi:
-$$T_i = \beta_0 + \beta_1 x_{i1} + ...+ \beta_{k}x_{ik}+ \epsilon_i= \beta_0 + \sum_{i=1}^k \beta_i x_{ij} + \epsilon_i$$

- $\epsilon$ - v.c. dell errore: errore campionario e l'errore dovuto alla mancvanza di ulteriori variabili esplicative.

Ipotesi:
- (a)- $E(\epsilon_i)= 0$ per ogni i
- (b1) - $Var)\epsilon_i= \sigma^2$ - ipotesi di omoschedasticita
- (b2) - $Cov(\epsilon_i, \epsilon_s)= 0; e\ne s$- ipotesi di incorrelazione
- (b1) e(b2) formano insieme - ipotesi di sfericita degli errori
- (c) - $X_j$ a natura deterministica e $k+1 < n$
- (d) - $\epsilon_i \sim N(0, \sigma^2)$ data (b2) le v.c. $\epsilon_i$ IID

Da (a) il modello si specifica: $$E[Y_i|x_{i1}...x_{ik}] = \beta_0 + \sum_{i=1}^k \beta_j x_{ij}$$

- questo e' un modello di regressione lineare

Riespressione delle ipotesi:
- (a) - $E(Y_i)= \mu_i$
- (b1) - $Var(Y_i)= \sigma^2$
- (b2) - $Cov(Y_i, Y_s)= 0$
- (d) - $Y \sim N(\mu_i, \sigma^2)$ - ora independeti ma NON identicamente distribuite, poiche $\mu_i \ne \mu_s$ per $i \ne s$

## Stima dei parametri del modello: risultati principali

1. Criterio dei Minimi Quadrati:
- vettore i cui elementi rendono minima la somma dei quarati dei reidui del modello $y=X\beta + e$

$$min_{\beta} \sum_{i=1}^n e_i^2= min_{\beta} e^t e= min_{\beta}(y-\hat{y})^t(y-\hat{y})$$

OLS- ordinari Least Squares - $\hat{\beta}^{OLS}$ per $\beta$ e la soluzione al sistema:
$$\hat{\beta^{OLS}}= (X^tX)^{-1}X^t y$$



2. Criterio di Massima Verosimiglianza:
- $L(\beta, \sigma^2, y)$

$$\hat{\beta}^{ML}= (X^tX)^{-1}X^t y$$

- coincide con la stima OLS. Pertano, sotto ipotesi distributiva normale due stime:
$$\hat{\beta}^{ML} == \hat{\beta}^{OLS}$$
$$\hat{B}^{ML} == \hat{B}^{OLS}$$
- la stime dei min.quadrati - $\hat{\beta}$
- stimatore- $\hat{B}$

Proprieta dello stimatore dei minimi quadrati:
- da (a) e (c)

1. Correttezza: $E(\hat{B})= \beta$
2. Linearita:
3. Stimatore OLS e' BLUE(Best Linear Unbiased Estimator) per vettore $\beta$


Altre conseguenze derivanti dalla procedura di stima dei  min.quadrati:
1. Nell Ipotesi (a') il mod.lineare
$$E(Y|X)= X\beta$$

$\hat{y}= X\hat{\beta}$ e' Stima del vettore dei valori attessi condizionati

- $\hat{e}= y-\hat{y}$ - residui dei minimi qudrati (residui OLS) esprimo gli scostamenti dei valori previsti
- $\hat{y}_i$ - fitted values
- $y_i$ - observed values


2. I residui OLS hanno somma (media) nulla

3. I residui OLS e i regressori $X_j$ sono fra loro incorrelati.

4. La somma dei qudrati dei residui OLS e' Deviazione Residua - SSE - SUm of Squares of Errors

$$SSE = \sum_{i=1}^n \hat{e}_i^2= \sum_{i=1}^n (y_i - \hat{y_i}^2)$$

5. Stima della varianza d'errore:
- ingenerale $\sigma^2$ e' ignota.


6. Scomposizione della Devianza Totale:

$$SST = \sum_{i=1}^n (y_i - \bar{y})^2$$

SST = SSE + SSR

- SSE - varianza residua
- SSR- devianza di regressione (Devianza SPiegata dal modello)

7. Indice di bonta di adattamento:

Indice di determinazione multipla:"
$$R^2 = \frac{SSR}{SST}= 1 - \frac{SSE}{SST}$$

- [0,1]
- $R^2= 0 <-> SSE=SST$ - il modello non spiega nulla dalla variabilita di Y
- $R^2=1 <-> SSR = SST$- quando il modello riproduce perfettamente i valori osservati di Y
